### 1. Dataset Overview
You will be working with a dataset related to retail sales. The dataset contains information about daily sales, product categories, store locations, and other relevant attributes.

**Dataset Name:** `retail_sales.csv`

**Attributes:**
- `Date`: The date of the sale.
- `Store`: The store where the sale occurred.
- `Product_Category`: The category of the product sold.
- `Sales_Amount`: The dollar amount of the sale.
- `Units_Sold`: The number of units sold.
- `Discount_Applied`: Whether a discount was applied (Yes/No).

```python
# Import necessary packages
# %pip install --upgrade pip
%pip install pandas
%pip install matplotlib
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('retail_sales.csv')

# Inspect the first few rows
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Generate summary statistics
print(df.describe())
```

### 3. Data Visualization
**Objective:** Create visualizations to uncover patterns and insights.

**Tasks:**
1. **Sales Over Time:** Create a line plot showing total sales over time.
2. **Sales by Store:** Create a bar chart comparing total sales across different stores.
3. **Product Category Analysis:** Create a pie chart showing the distribution of sales across different product categories.
4. **Impact of Discounts:** Use a boxplot to compare the sales amounts for transactions with and without discounts.

```python
import matplotlib.pyplot as plt

# Sales over time
df.groupby('Date')['Sales_Amount'].sum().plot(kind='line', title='Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.show()

# Sales by Store
df.groupby('Store')['Sales_Amount'].sum().plot(kind='bar', title='Total Sales by Store')
plt.xlabel('Store')
plt.ylabel('Total Sales')
plt.show()

# Product Category Analysis
df.groupby('Product_Category')['Sales_Amount'].sum().plot(kind='pie', autopct='%1.1f%%', title='Sales by Product Category')
plt.ylabel('')
plt.show()

# Impact of Discounts
df.boxplot(column='Sales_Amount', by='Discount_Applied')
plt.title('Sales Amount by Discount Applied')
plt.suptitle('')
plt.xlabel('Discount Applied')
plt.ylabel('Sales Amount')
plt.show()
```
